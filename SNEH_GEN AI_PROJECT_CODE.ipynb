```python
import pandas as pd

# Load the new dataset from the provided path using the python engine
df_amazon = pd.read_csv("/content/amazon.csv", engine='python')

# Display the first few rows and the columns and their data types
display(df_amazon.head())
display(df_amazon.info())
```

```python
from apyori import apriori

# Apply the Apriori algorithm
# Adjusted parameters to find more rules
rules = apriori(transactions_amazon, min_support=0.001, min_confidence=0.1, min_lift=1.1, min_length=2)

# Convert the rules to a list
results = list(rules)

# Display the first 10 rules
print("First 10 association rules:")
for i, rule in enumerate(results[:10]):
    print(f"Rule {i+1}: {rule}")
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Extract the antecedent and consequent from the first rule
# Assuming 'results' list is not empty and contains at least one rule
if results:
    first_rule = results[0]
    for ordered_stat in first_rule.ordered_statistics:
        antecedent = list(ordered_stat.items_base)
        consequent = list(ordered_stat.items_add)
        confidence = ordered_stat.confidence # Get confidence here
        break # Take the first ordered statistic

    if antecedent and consequent:
        # Create a directed graph
        G = nx.DiGraph()

        # Add nodes for antecedent and consequent
        antecedent_node = antecedent[0] if antecedent else "Antecedent"
        consequent_node = consequent[0] if consequent else "Consequent"

        G.add_node(antecedent_node)
        G.add_node(consequent_node)

        # Add an edge representing the association
        G.add_edge(antecedent_node, consequent_node, weight=confidence) # Use confidence directly

        # Draw the graph
        pos = nx.spring_layout(G) # Layout for better visualization
        plt.figure(figsize=(12, 8)) # Increased figure size for better readability
        nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=10, font_weight="bold", arrows=True, arrowstyle='->', arrowsize=20)

        # Manually add edge labels for confidence
        edge_labels = {(antecedent_node, consequent_node): f"Confidence: {confidence:.2f}"}
        nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red', font_size=10)

        plt.title("Association Rule Visualization")
        plt.show()
    else:
        print("Could not extract antecedent and consequent from the rule.")
else:
    print("No association rules found to visualize.")
```

```python
# Function to get cross-selling suggestions for a given product
def get_cross_selling_suggestions(product_name, rules_list):
    suggestions = []
    for rule in rules_list:
        for ordered_stat in rule.ordered_statistics:
            # Check if the input product is in the antecedent of the rule
            if product_name in ordered_stat.items_base:
                # Add the consequent products as suggestions
                suggestions.extend(ordered_stat.items_add)
    # Return unique suggestions
    return list(set(suggestions))

# Example usage: Get suggestions for the pTron Micro USB Cable
product_to_suggest_for = 'pTron Solero MB301 3A Micro USB Data & Charging Cable, Made in India, 480Mbps Data Sync, Strong & Durable 1.5-Meter Nylon Braided USB Cable for Micro USB Devices - (Black)'
suggestions = get_cross_selling_suggestions(product_to_suggest_for, results)

print(f"Cross-selling suggestions for '{product_to_suggest_for}':")
if suggestions:
    for suggestion in suggestions:
        print(f"- {suggestion}")
else:
    print("No suggestions found for this product based on the current rules.")

# Example usage: Get suggestions for a product not in the rules
product_to_suggest_for_2 = 'Bajaj Rex 500W Mixer Grinder with Nutri-Pro Feature, 3 Jars, White'
suggestions_2 = get_cross_selling_suggestions(product_to_suggest_for_2, results)

print(f"\nCross-selling suggestions for '{product_to_suggest_for_2}':")
if suggestions_2:
    for suggestion in suggestions_2:
        print(f"- {suggestion}")
else:
    print("No suggestions found for this product based on the current rules.")
```

```python
# Function to extract and print rules in a readable format
def inspect_rules(results):
    for rule in results:
        # Extract items from the rule
        items = [x for x in rule.items]

        # Extract support, confidence, and lift
        support = rule.support

        # Extract antecedent and consequent from ordered statistics
        for ordered_stat in rule.ordered_statistics:
            antecedent = [x for x in ordered_stat.items_base]
            consequent = [x for x in ordered_stat.items_add]
            confidence = ordered_stat.confidence
            lift = ordered_stat.lift

            print(f"Rule: {antecedent} -> {consequent}")
            print(f"Support: {support:.4f}")
            print(f"Confidence: {confidence:.4f}")
            print(f"Lift: {lift:.4f}")
            print("-" * 30)

# Inspect all the rules found
print("All found association rules:")
inspect_rules(results)
```
